# Visualizaing data is essential and presenting data.
# The saying "A picture is worth a thousand words" certainly applies to understanding data.



```python
import matplotlib.pyplot as plt # standard import for plotting

X = [ 0, 1, 2, 3, 4 ] 
Y = [ 20, 25, 35, 50, 10]

plt.plot(X, Y) # plot X and Y using default line style and color
```

```python
# Styling the plot

X = [ 0, 1, 2, 3, 4 ] 
Y = [ 20, 25, 35, 50, 10]

plt.plot(X, Y, marker = 's') # plot X and Y using default line style and color
```

```python
# Styling the plot

X = [  0,   1,  2,  3,  4] 
Y = [ 20,  25, 35, 50, 10]

plt.plot(X, Y, marker = 's', linestyle = ':')
```
